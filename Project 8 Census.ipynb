### Q1 — How to hide the index of the DataFrame?

> **Task:** Use `style.hide_index()` to display the DataFrame without showing the default numeric index.

> 💡 **Why?** When presenting data to stakeholders or in reports, the index (0, 1, 2...) adds visual noise without meaning.  
> Hiding it makes the table cleaner and easier to read.

### Q4 — State-wise Total Population & Religion-wise Population

> **Task:**  
> (A) Calculate the total population per state using `groupby().sum()`  
> (B) Calculate the total count of each religion per state


```python
# set_caption() adds a title above the DataFrame
# .hide() removes the index — combined for a clean professional display
data.head(10).style.set_caption("India Census 2011 — District-Level Data").hide()
```

```python
# Religion-wise population per state
religion_cols = ['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']
state_religion = data.groupby('State_name')[religion_cols].sum()

print("State-wise Religion Distribution:")
state_religion.style.set_caption("State-wise Religion Distribution — India Census 2011")
```

### Q5 — How many Male Workers were there in Maharashtra?

> **Task:** Filter for Maharashtra using `df[df.Col == 'Element']['Col2']`, then sum Male_Workers.

> 💡 **Two approaches** — both give the same answer:  
> - Filter first, then `.sum()` directly  
> - Use `groupby()` to get the answer for ALL states at once and then index into Maharashtra


```python
# Approach 1: Direct filtering — isolate Maharashtra rows, then sum Male_Workers
maharashtra_male_workers = data[data['State_name'] == 'MAHARASHTRA']['Male_Workers'].sum()
print(f"Total Male Workers in Maharashtra: {maharashtra_male_workers:,}")
```

```python
# Approach 2: groupby — useful when you want Male Workers for ALL states at once
all_state_male_workers = data.groupby('State_name')['Male_Workers'].sum()

# Index into Maharashtra specifically
print(f"Maharashtra (via groupby): {all_state_male_workers['MAHARASHTRA']:,}")

# Top 10 states by Male Workers
print("\nTop 10 States by Male Workers:")
print(all_state_male_workers.sort_values(ascending=False).head(10))
```

### Q6 — How to set a column as the index of the DataFrame?

> **Task:** Use `set_index()` to replace the default numeric index with a meaningful column.

> 💡 **Why set_index()?**  
> Setting District_name as index allows label-based lookup using `.loc['District']`  
> instead of searching through the whole DataFrame.  
> It also makes groupby and merge operations more intuitive.


```python
# Set 'District_name' as the index — makes label-based access much faster
data_indexed = data.set_index('District_name')

print("Index name:", data_indexed.index.name)
data_indexed.head()
```

```python
# Now we can access a district directly by name using .loc[]
data_indexed.loc['New Delhi']
```

```python
# Reset index back to default if needed for further analysis
data_reset = data_indexed.reset_index()
print("Back to default index:", data_reset.index.tolist()[:5])
```

### Q7a & Q7b — Add Suffix and Prefix to Column Names

> **Task:**  
> (7a) Add a suffix to all column names using `add_suffix()`  
> (7b) Add a prefix to all column names using `add_prefix()`

> ⚠️ **Important:** `add_suffix()` and `add_prefix()` return a **new DataFrame** — they don't modify in place.  
> Always assign the result to a variable or use `inplace` equivalent patterns.  
> These are especially useful when merging multiple datasets to avoid column name conflicts.


```python
# Q7a: add_suffix() — appends '_2011' to every column name
# Useful when merging multiple census years to distinguish columns
data_suffix = data.add_suffix('_2011')

print("Columns after add_suffix('_2011'):")
print(data_suffix.columns.tolist())
data_suffix.head(3)
```

```python
# Q7b: add_prefix() — prepends 'census_' to every column name
# Useful when merging with non-census datasets to clarify data source
data_prefix = data.add_prefix('census_')

print("Columns after add_prefix('census_'):")
print(data_prefix.columns.tolist())
data_prefix.head(3)
```

## 7. Visualizations

### 7.1 Univariate — Numerical Variables

```python
# Distribution of Population, Male, Female across all districts
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, col, color in zip(axes,
                           ['Population', 'Male', 'Female'],
                           ['steelblue', 'cornflowerblue', 'lightcoral']):
    sns.histplot(data[col], kde=True, ax=ax, color=color)
    ax.set_title(f'Distribution of {col}')

plt.suptitle("Population Distribution — India Census 2011", y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Skewness — population data is typically right-skewed (a few very large districts)
print("Skewness:")
print(data[['Population', 'Male', 'Female', 'Workers']].skew().round(2))
```

```python
# Boxplots for outlier detection
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, col, color in zip(axes,
                           ['Population', 'Workers', 'Literate'],
                           ['lightblue', 'lightgreen', 'lightyellow']):
    sns.boxplot(x=data[col], ax=ax, color=color)
    ax.set_title(f'Boxplot — {col}')

plt.suptitle("Outlier Detection — India Census 2011", y=1.02)
plt.tight_layout()
plt.show()
```

### 7.2 Top States — Bar Charts

```python
# Top 10 states by total population
top10_pop = data.groupby('State_name')['Population'].sum().nlargest(10)

plt.figure(figsize=(10, 6))
top10_pop.sort_values().plot(kind='barh', color='steelblue')
plt.title("Top 10 States by Total Population")
plt.xlabel("Total Population")
plt.tight_layout()
plt.show()
```

```python
# Top 10 states by Male Workers
top10_male_workers = data.groupby('State_name')['Male_Workers'].sum().nlargest(10)

plt.figure(figsize=(10, 6))
top10_male_workers.sort_values().plot(kind='barh', color='cornflowerblue')
plt.title("Top 10 States by Male Workers")
plt.xlabel("Male Workers")
plt.tight_layout()
plt.show()
```

### 7.3 Religion Distribution

```python
# National-level religion breakdown as a pie chart
religion_cols = ['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']
national_religion = data[religion_cols].sum()

plt.figure(figsize=(8, 8))
plt.pie(national_religion,
        labels=national_religion.index,
        autopct='%1.1f%%',
        startangle=140,
        wedgeprops=dict(linewidth=0.5, edgecolor='white'))
plt.title("National Religion Distribution — India Census 2011", pad=20)
plt.tight_layout()
plt.show()
```

```python
# Religion distribution for Top 5 most populous states
top5_states = data.groupby('State_name')['Population'].sum().nlargest(5).index
religion_top5 = data[data['State_name'].isin(top5_states)].groupby('State_name')[religion_cols].sum()

religion_top5.plot(kind='bar', figsize=(12, 6), stacked=True)
plt.title("Religion Distribution — Top 5 Most Populous States")
plt.xlabel("State")
plt.ylabel("Population")
plt.xticks(rotation=15)
plt.legend(title='Religion', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

### 7.4 Bivariate — Numerical vs Numerical

```python
# Population vs Workers — more people = more workers?
sns.regplot(x='Population', y='Workers', data=data,
            scatter_kws={'alpha': 0.4}, line_kws={'color': 'red'})
plt.title("Population vs Workers (Regression Line)")
plt.show()
```

```python
# Population vs Literate — literacy rate vs population size
sns.scatterplot(x='Population', y='Literate', data=data, alpha=0.5, color='steelblue')
plt.title("Population vs Literate Population")
plt.show()
```

### 7.5 Age Group Distribution

```python
# National-level age group breakdown
age_cols = ['Age_Group_0_29', 'Age_Group_30_49', 'Age_Group_50']
national_age = data[age_cols].sum()

plt.figure(figsize=(7, 5))
national_age.plot(kind='bar', color=['steelblue', 'seagreen', 'coral'])
plt.title("National Age Group Distribution — India Census 2011")
plt.xlabel("Age Group")
plt.ylabel("Population")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### 7.6 Correlation Heatmap

```python
# Select key numerical columns for correlation analysis
key_cols = ['Population', 'Male', 'Female', 'Literate', 'Workers',
            'Male_Workers', 'Female_Workers', 'Hindus', 'Muslims']

sns.heatmap(data[key_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap — India Census 2011")
plt.tight_layout()
plt.show()
```

### 7.7 Pairplot

```python
# Pairplot of key population variables
sns.pairplot(data[['Population', 'Literate', 'Workers', 'Male_Workers', 'Female_Workers']])
plt.suptitle("Pairplot — India Census 2011", y=1.02)
plt.show()
```

## 8. Key Insights

After analyzing **640 districts across 35 Indian states — India Census 2011**:

1. **Uttar Pradesh is the most populous state**  
   State-wise groupby reveals UP leads with the highest total population, followed by Maharashtra and Bihar — together accounting for a disproportionate share of India's total population.

2. **Hinduism is the dominant religion nationally (~80%)**  
   The pie chart of national religion distribution confirms Hinduism dominates, with Islam being the second-largest religion and Christianity, Sikhism, Buddhism, and Jainism making up the remainder.

3. **Maharashtra has the highest number of Male Workers**  
   With 32,616,875 male workers, Maharashtra leads all states — reflecting its large industrial and manufacturing workforce.

4. **Population and Workers are very strongly correlated**  
   As expected, districts with larger populations almost always have more workers — the regression line is steep and tight, confirming a near-linear relationship.

5. **India's population is predominantly young (0–29 age group)**  
   The age group bar chart shows that the 0–29 age group forms the largest segment nationally, consistent with India's demographic dividend profile in 2011.

6. **Female workers are significantly fewer than Male workers across all states**  
   The correlation heatmap and pairplot reveal that while Male_Workers and Female_Workers are both correlated with Population, Female_Workers consistently show a much lower count — highlighting gender disparities in workforce participation.
